# What is linear regression?
According to Anon. n.d., Linear regression is a statistical method used for forecasting the value of one variable based on another. The variable, X, to be predicted is termed the dependent variable, while the variable, Y, used for prediction is known as the independent variable. This analytical technique involves determining the coefficients of a linear equation, incorporating one or more independent variables to forecast the dependent variable’s value most accurately. Linear regression establishes a straight line or that minimizes the disparities between predicted and actual output values. Various tools, such as simple linear regression calculators, employ the "least squares" approach to identify the optimal-fit line for a given dataset, allowing for the estimation of the dependent variable's value from the independent variable. Linear regression models offer a straightforward and interpretable mathematical framework for prediction, making them widely applicable across business and academic domains. According to Anon. n.d., linear regression is used in diverse fields ranging from biological, behavioral, environmental, and social sciences to business contexts. Due to their established efficacy, linear regression models serve as a trusted method for scientifically and reliably forecasting future outcomes. The longstanding use of linear regression has led to a comprehensive understanding of its properties, facilitating swift training and implementation. Anon. n.d., states that Linear regression is a statistical method used in data science and machine learning for predictive analysis. Its primary objective revolves around addressing two key inquiries: “how effectively can a set of predictor variables forecast an outcome (dependent or criterion) variable and which specific variables emerge as significant predictors of the outcome variable, and how do their beta estimates—reflecting both magnitude and direction—affect this outcome”. 

The simplest form of linear regression involves one dependent variable and one independent variable, encapsulated within the equation 
y = c + b*x, where:
y represents the predicted score of the dependent variable,
c denotes the constant,
b signifies the regression coefficient, and
x represents the score on the independent variable.
The y-intercept is the predicted value for the response (y) when x = 0. The slope describes the change in y for each one unit change in x.

Multiple Linear Regression involves more than one independent variable and one dependent variable. The equation for multiple linear regression is: y = β0+β1X+β2X+………βnX where:
Y is the dependent variable
X1, X2, …, Xp are the independent variables
β0 is the intercept
β1, β2, …, βn are the slopes

According to Gupta, M. n.d., Linear regression, categorized as a supervised machine learning algorithm, operates by learning from labelled datasets and establishing optimized linear functions that map data points. Through this process, it discerns patterns within the data and creates a linear relationship between input variables and the target variable. This trained model can then be employed to make predictions on new datasets by extrapolating the learned patterns to unseen data points. The main aim is to identify the best-fit line, minimizing the discrepancy between predicted and actual values. The best-fit line is characterized by the least amount of error. This line is described by an equation that represents the relationship between the dependent and independent variables. The slope of this line signifies the extent of change in the dependent variable for a unit alteration in the independent variable(s).

Kanade, V. 2023., states that linear regression provides the following advantages:
1.	Simplified Implementation: Linear regression's computational simplicity translates to easy implementation, requiring minimal engineering overhead before and during model deployment.
2.	Interpretability: Unlike complex deep learning models such as neural networks, linear regression offers straightforward interpretability. This transparency is advantageous over black-box models, as it facilitates understanding which input variables directly influence changes in the output variable.
3.	Scalability: Linear regression's lightweight computational requirements make it well-suited for scalable applications, particularly in scenarios with large datasets or big data environments.
4.	Suitability for Online Settings: The computational efficiency of linear regression enables its use in online settings, allowing for real-time training and prediction generation. This contrasts with computationally intensive models like neural networks or support vector machines, which necessitate significant computing resources and time for retraining on new data. Therefore, linear regression proves cost-effective and practical for real-time applications.

Gupta, M. n.d., states that linear regression has the following disadvantages:
1.	Linear regression operates under the assumption of a linear relationship between the dependent and independent variables. Should this relationship not hold true, the model's performance may not do well.
2.	Linear regression is susceptible to multicollinearity, where high correlations between independent variables can inflate coefficient variances and destabilize predictions.
3.	Linear regression assumes that features are appropriately formatted for the model's use, potentially necessitating feature engineering to ensure effective utilization.

```python
# 1: Scatter plot graph to visualize the relationship between hours of study and test scores.

#Import Required Libraries
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.tree import DecisionTreeRegressor
from sklearn.datasets import make_regression

# Load data from CSV file
data = pd.read_csv(r"C:\Users\lab_services_student\archive\student_scores.csv")

# Extract hours and scores from the dataframe
hours = data['Hours']
scores = data['Scores']

# Scatter plot 
plt.figure(figsize=(10, 6))
plt.scatter(hours, scores, color='blue', alpha=0.7)
plt.title('Scatter plot showing the relationship between Hours and Scores')
plt.xlabel('Hours')
plt.ylabel('Scores')
plt.grid(True)

# Customize x-axis to show every 0.5 hours
custom_ticks_x = np.arange(0, max(hours) + 1, 0.5)  
plt.xticks(custom_ticks_x)

# Customize y-axis to increment in 10s
custom_ticks_y = np.arange(0, max(scores) + 1, 10)  
plt.yticks(custom_ticks_y)

plt.show()

# Code Attribution:
# Anon. n.d. Matplotlib Scatter. W3schools. [Online] Available at: https://www.w3schools.com/python/matplotlib_scatter.asp [Accessed 05 April 2024]
# Anon. n.d. Machine Learning - Scatter Plot. W3schools. [Online] Available at: https://www.w3schools.com/python/python_ml_scatterplot.asp  [Accessed 05 April 2024]
```

```python
# 2 and 3 Create scatter plot with regression line for training data and testing data

# Reshape data for sklearn LinearRegression
Hours_reshaped = hours.values.reshape(-1, 1)
Scores_reshaped = scores.values.reshape(-1, 1)

# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(Hours_reshaped, Scores_reshaped, test_size=0.2, random_state=42)

# Create and fit Linear Regression model
model = LinearRegression()
model.fit(X_train, y_train)

# Predict scores
predicted_train_scores = model.predict(X_train)
predicted_test_scores = model.predict(X_test)

# Define x-axis range and tick marks
x_range = np.arange(0, 10.5, 0.5)

# Define y-axis range and tick marks
y_range = np.arange(0, 101, 10)

# 2: Create scatter plot with regression line for training data
plt.figure(figsize=(10, 6))
plt.scatter(X_train, y_train, color='blue', alpha=0.7, label='Training Data')
plt.plot(X_train, predicted_train_scores, color='red', label='Regression Line')
plt.title('Scatter plot showing linear regression analysis to model relationship between Hours and Scores (Training Data)')
plt.xlabel('Hours')
plt.ylabel('Scores')
plt.xticks(x_range)
plt.yticks(y_range)
plt.xlim(0, 10)
plt.ylim(0, 100)
plt.grid(True)

# Add legend outside the plot area
plt.legend(loc='upper center', bbox_to_anchor=(0.5, -0.1), fancybox=True, shadow=True, ncol=2)

# Evaluate model on training data
mse_train = mean_squared_error(y_train, predicted_train_scores)
r_squared_train = r2_score(y_train, predicted_train_scores)

# Print evaluation metrics for training data
print("Training Data - Mean Squared Error:", mse_train)
print("Training Data - R-squared:", r_squared_train)

plt.show()

# 3: Create scatter plot with regression line for testing data
plt.figure(figsize=(10, 6))
plt.scatter(X_test, y_test, color='blue', alpha=0.7, label='Testing Data')
plt.plot(X_test, predicted_test_scores, color='red', label='Regression Line')
plt.title('Scatter plot showing linear regression analysis to model relationship between Hours and Scores (Testing Data)')
plt.xlabel('Hours')
plt.ylabel('Scores')
plt.xticks(x_range)
plt.yticks(y_range)
plt.xlim(0, 10)
plt.ylim(0, 100)
plt.grid(True)

# Add legend outside the plot area
plt.legend(loc='upper center', bbox_to_anchor=(0.5, -0.1), fancybox=True, shadow=True, ncol=2)

# Evaluate model on testing data
mse_test = mean_squared_error(y_test, predicted_test_scores)
r_squared_test = r2_score(y_test, predicted_test_scores)

# Print evaluation metrics for testing data
print("Testing Data - Mean Squared Error:", mse_test)
print("Testing Data - R-squared:", r_squared_test)

plt.show()

# Code Attribution:
# Anon. n.d. Machine Learning - Linear Regression. W3schools. [Online] Available at: https://www.w3schools.com/python/python_ml_linear_regression.asp [Accessed 05 April 2024]
# Anon. 2023. Linear Regression (Python Implementation). GeeksforGeeks. [Online] Available at: https://www.geeksforgeeks.org/linear-regression-python-implementation/ [Accessed 05 April 2024]
```

# 5: Residual plot to plot residuals based on actual vs predicted scores
- According to Anon. n.d., The residual plot is a graphical representation of the differences between the observed test scores and the predicted test scores. It is a fundamental tool for assessing the performance and validity of a regression model
- The x-axis represents the predicted scores for each observation based on the regression model. Each point on the plot corresponds to a specific prediction made by the model.
- The y-axis represents the residuals. The residuals are the differences between the observed scores and the predicted scores. Each point on the plot represents the residual for a particular observation.  They are the vertical distances between the actual data points and the regression line and is calculated as the observed score minus the predicted score.
- Red Dashed Line at y=0 indicates the ideal scenario where residuals are exactly zero, meaning the predicted scores perfectly match the observed scores.
- Points above this line indicate that the model underpredicted the scores, while points below the line indicate overprediction.
- By examining the spread and distribution of the residuals, we can identify any patterns or heteroscedasticity. Heteroscedasticity refers to the situation where the spread of the residuals varies across the range of predicted values. If there's a clear pattern or if the spread of the residuals changes systematically across the range of predicted values, it indicates that the model may not be capturing all the relevant information in the data. This can help identify potential issues with the regression model, such as incorrect functional form or missing predictor variables.
- In this case, residuals are consistently above the red line (which represents the y-axis at y=0), which indicates underfitting. The regression model tends to underpredict the scores across all levels of study hours. The model systematically predicts lower scores than the actual observed scores. this supports the assumption that the model does not be capture all the relevant factors that influence test scores. For instance, factors like student motivation, prior knowledge, or individual differences in learning styles could be missing from the model.
- To combat this issue, we will incorporate a decision tree to provide a more accurate representation. 

```python
#6 Residual plot for actual vs predicted scores using decision tress and regularization

# Generate synthetic data for demonstration
X, y = make_regression(n_samples=100, n_features=1, noise=10, random_state=42)

# Shift the target variable y by a constant to ensure all values are positive
y_shifted = y - y.min() + 1  # Add 1 to ensure all values are greater than zero

# Apply logarithmic transformation to the shifted target variable
y_log = np.log1p(y_shifted)

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y_log, test_size=0.2, random_state=42)

# Create and fit Decision Tree model with regularization
tree_model = DecisionTreeRegressor(max_depth=5, min_samples_split=5, min_samples_leaf=2)
tree_model.fit(X_train, y_train)

# Predict log-transformed scores for the testing set using the decision tree model
predicted_scores_tree_log = tree_model.predict(X_test)

# Convert back to original scale by exponentiations
predicted_scores_tree = np.expm1(predicted_scores_tree_log)

# Scale predicted scores to the range (0, 100)
predicted_scores_tree_scaled = (predicted_scores_tree - predicted_scores_tree.min()) / (predicted_scores_tree.max() - predicted_scores_tree.min()) * 100

# Calculate residuals
residuals_tree = np.expm1(y_test) - predicted_scores_tree

# Plot residuals against predicted scores using the decision tree model
plt.scatter(predicted_scores_tree_scaled, residuals_tree, color='blue', alpha=0.7, label='Residuals')
plt.title('Residual Plot for Actual Scores vs Predicted Scores (Decision Tree)')
plt.xlabel('Predicted Scores (Decision Tree)')
plt.ylabel('Residuals')
plt.axhline(y=0, color='r', linestyle='--', label='Zero Residual Line')  # Adding a horizontal line at y=0 for reference
plt.legend(loc='upper center', bbox_to_anchor=(0.5, -0.1), shadow=True, ncol=2)

plt.grid(True)
plt.show()

# Code Attribution:
# Navlani, A. 2023. Decision Tree Classification in Python Tutorial. DataCamp. [Online] Available at: https://www.datacamp.com/tutorial/decision-tree-classification-python [Accessed 07 April 2024]
```

# 6: Residual plot for actual vs predicted scores using decision tree and regularization
- X-axis (Predicted Scores): This axis represents the predicted scores for the testing set obtained from the decision tree model. The predicted scores have been scaled to the range of 0 to 100.
- Y-axis (Residuals): This axis represents the residuals, which are the differences between the actual scores and the predicted scores. A residual is calculated for each data point in the testing set.
- Blue Dots: Each blue dot on the plot represents a data point in the testing set. The position of the dot along the X-axis indicates the predicted score for that data point, and the position along the Y-axis indicates the corresponding residual.
- Zero Residual Line (Red Dashed Line): The red dashed line at y=0 serves as a reference line. It indicates the zero residual line, where the residuals equal zero. Data points above this line have positive residuals (actual score > predicted score), while data points below the line have negative residuals (actual score < predicted score).
- Regularization: Regularization is a technique used to prevent overfitting by adding a penalty to the loss function used to train the model. Regularization aims to make the model simpler and more generalizable, but its effects would be seen in the model's performance metrics rather than directly observed in this residual plot.
- The residuals are spread out across different predicted scores, which suggests variability in the accuracy of the predictions. There is a slight concentration of data points with higher residuals as the predicted scores increase. The residuals on this plot seem to be distributed somewhat randomly around the zero residual line, without any clear pattern or systematic structure. This suggests that the decision tree has not overfitted to a particular subset of the data, which is good. Overfitting would be indicated by a clear pattern or trend in the residuals. There is no clear pattern that suggests systematic errors in the model; the residuals appear to be randomly scattered, which is a good sign in residual analysis.
- Some residuals are quite large (both positive and negative), particularly for higher predicted scores. This could indicate that the decision tree model has more difficulty accurately predicting higher scores or that there is more variance in the actual scores that the model is not capturing. The presence of some large residuals indicates that there might be room for improvement.
- There are both positive and negative residuals, which means the model sometimes overestimates and sometimes underestimates the actual scores. There does not appear to be a bias towards overestimation or underestimation across the range of predicted scores, which is also an indication of a well-performing model.
